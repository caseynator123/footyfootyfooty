## Go get weather data

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.set_window_size(1200,900)
```

```python
df = pd.DataFrame({},columns=['location','year','month','day','time','temp','dew point','humidity','wind','wind speed','wind gust','pressure','precip','condition'])
```

```python
import time
```

```python
def get_daily_weather(location, weather_station, year, month, day):
    driver.get(f'https://www.wunderground.com/history/daily/{weather_station}/date/{year}-{month}-{day}')
    time.sleep(0.5)
    for i in range(1,100):
        try:
            row = driver.find_element_by_xpath(f'(//*[@class="mat-row cdk-row ng-star-inserted"])[{i}]').text.split()
            row = [s for s in row if s!='°F' and s!='mph' and s!='in' and s!='%']
            row[0:2] = [' '.join(row[0:2])]
            if len(row) > 10:
                row[9:] = [' '.join(row[9:])]
            row = [location, year, month, day] + row
            print(row)
            df.loc[len(df)] = row
        except:
            break
```

```python
for key in venue.keys():
    season_data.loc[season_data['Venue']==key, 'location'] = venue[key]
for key in weather_station.keys():
    season_data.loc[season_data['Venue']==key, 'weather_station'] = weather_station[key]
```

```python
season_data['year'] = season_data['year'].apply(str)
```

```python
season_data
```

```python
season_data['Date'] = season_data['Date'].str.extract(r'(\d{1,2} \w{3})')
```

```python
season_data['Date'] = season_data['Date'] + ' ' + season_data['year']
```

```python
from datetime import datetime
```

```python
def rearrange_date(x):
    return datetime.strftime(datetime.strptime(x, '%d %b %Y'), '%-d %-m %Y')
```

```python
season_data['Date'] = season_data['Date'].apply(lambda x: rearrange_date(x))
```

```python
season_data['day'] = season_data['Date'].apply(lambda x: x.split()[0])
season_data['month'] = season_data['Date'].apply(lambda x: x.split()[1])
season_data['year'] = season_data['Date'].apply(lambda x: x.split()[2])
```

```python
season_data = season_data[['day','month','year','location','weather_station']].drop_duplicates().reset_index(drop=True)
```
