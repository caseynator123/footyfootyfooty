```python
df = pd.DataFrame({})
for season in sorted(os.listdir('/Users/caseyh/Desktop/footyfootyfooty/round_results/'))[1:-1]:
    tmp = pd.read_csv(f'/Users/caseyh/Desktop/footyfootyfooty/round_results/{season}')
    df = pd.concat([df,tmp])
df = df.reset_index(drop=True)
```

```python
df.loc[df['Home']=='Fitzroy ', 'Home'] = 'FIT'
df.loc[df['Away']==' Fitzroy','Away'] = 'FIT'
```

```python
# change round values to ints
df['round'] = df['round'].str[1:].astype(int)
```

```python
# add in bye rounds:
for year in range(1990,2024):
    for team in df[df['year']==year]['Home'].unique():
        for r in df[df['year']==year]['round'].unique():
            if r not in list(df[(df['Home'] == team)&(df['year']==year)]['round']) \
            and r not in list(df[(df['Away'] == team)&(df['year']==year)]['round']):
                new_row = pd.DataFrame({'Home':team, 'round':r,'Away':'BYE', 'year':year}, index=[0])
                df = pd.concat([new_row,df]).reset_index(drop=True)
df = df.sort_values(['year', 'round']).reset_index(drop=True)
```

```python
df['actual_outcome'] = df['Home Score']/(df['Home Score']+df['Away Score'])
```

```python
def get_new_rating(team_elo, k, actual_outcome, expected_outcome):
    return team_elo + k * (actual_outcome - expected_outcome)
def get_expected_outcome(team_elo, opp_elo, hga, s):
    return 1/(1+10**((opp_elo-team_elo-hga)/s))
```

```python
elo_table = pd.DataFrame({}, columns=['team','year','round','team_elo'])

# start season
for year in range(1990,2024):
    for team in df[df['year']==year]['Home'].unique():
        if team not in elo_table['team'].unique():
            new_row = pd.DataFrame({'team':team, 'round':0,'team_elo':1000,'year':year }, index=[0])
            elo_table = pd.concat([new_row,elo_table]).reset_index(drop=True)
```

```python
for year in range(1990,2024):
    
    rounds = df[df['year']==year]['round'].unique()
    
    print(year)
    for r in rounds:
        
        elo_rounds = elo_table[elo_table['year']==year]['round'].unique()
        
        if r in elo_rounds:

            elo_table = elo_table.drop(elo_table[(elo_table['year']==year)&(elo_table['round']==r)].index)
            
        for team in elo_table[(elo_table['year']==year)]['team'].unique():
            if df[(df['year']==year)&(df['Home'] == team)&(df['round']==r)]['Away'].shape[0]==0:
                continue

            if df[(df['year']==year)&(df['Home'] == team)&(df['round']==r)]['Away'].values[0] == 'BYE':
                tmp_elo = elo_table[(elo_table['year']==year)&(elo_table['team']==team)&(elo_table['round']==r-1)]['team_elo'].values[0]
                new_row = pd.DataFrame({'team':team,'round':r,'team_elo':tmp_elo,'year':year}, index=[0])
                elo_table = pd.concat([elo_table, new_row])
                if r == rounds.max():
                    if tmp_elo < 1000:
                        tmp_elo = tmp_elo * 1.05
                    else:
                        tmp_elo = tmp_elo * 0.95
                    new_row = pd.DataFrame({'team':team,'round':0,'team_elo':tmp_elo,'year':year+1}, index=[0])
                    elo_table = pd.concat([elo_table, new_row])
                continue
                              
                              
            opp = df[(df['year']==year)&(df['Home'] == team)&(df['round']==r)]['Away'].values[0]
            
            team_elo = elo_table[(elo_table['year']==year)&(elo_table['team'] == team)&(elo_table['round'] == r-1)]['team_elo'].values[0]
            opp_elo = elo_table[(elo_table['year']==year)&(elo_table['team'] == opp)&(elo_table['round'] == r-1)]['team_elo'].values[0]

            actual_outcome = df[(df['year']==year)&(df['Home'] == team)&(df['round']==r)]['actual_outcome'].values[0]

            expected_outcome = get_expected_outcome(team_elo, opp_elo, 20, 10000)

            n_team_elo = team_elo + 100 * (actual_outcome - expected_outcome)
            n_opp_elo = opp_elo - 100 * (actual_outcome - expected_outcome)

            new_rows = pd.DataFrame({'team':[team, opp],'round':[r,r],'team_elo':[n_team_elo,n_opp_elo],'year':[year,year]}, index=[0,1])
            
            elo_table = pd.concat([elo_table, new_rows])
            
            df.loc[(df['year']==year)&(df['Home']==team)&(df['round']==r),'expected_outcome'] = expected_outcome
    
            if r == rounds.max():
                if n_team_elo < 1000:
                        n_team_elo = n_team_elo * 1.05
                else:
                    n_team_elo = n_team_elo * 0.95
                
                if n_opp_elo < 1000:
                        n_opp_elo = n_opp_elo * 1.05
                else:
                    n_opp_elo = n_opp_elo * 0.95
                    
                new_row = pd.DataFrame({'team':[team,opp],'round':[0,0],'team_elo':[n_team_elo,n_opp_elo],'year':[year+1,year+1]}, index=[0,1])
                
                elo_table = pd.concat([elo_table, new_row])
    elo_table = elo_table.sort_values(['year','round', 'team_elo'], ascending=[True,True,False]).reset_index(drop=True)
```
