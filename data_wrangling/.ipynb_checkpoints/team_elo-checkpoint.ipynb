```python
df = pd.read_csv('/Users/caseyh/Desktop/footyfootyfooty/round_results/2023_round_results.csv')
```

```python
# change round values to ints
df['round'] = df['round'].str[1:].astype(int)
```

```python
# add in bye rounds:
for team in df['Home'].unique():
    for r in df['round'].unique():
        if r not in list(df[df['Home'] == team]['round']) and r not in list(df[df['Away'] == team]['round']):
            new_row = pd.DataFrame({'Home':team, 'round':r,'Away':'BYE'}, index=[0])
            df = pd.concat([new_row,df]).reset_index(drop=True)
df = df.sort_values('round').reset_index(drop=True)
```

```python
df['actual_outcome'] = df['Home Score']/(df['Home Score']+df['Away Score'])
```

```python
def get_new_rating(team_elo, k, actual_outcome, expected_outcome):
    return team_elo + k * (actual_outcome - expected_outcome)
def get_expected_outcome(team_elo, opp_elo, hga, s):
    return 1/(1+10**((opp_elo-team_elo-hga)/s))
```

```python
df.head(10)
```

```python
for r in df['round'].unique():
    if r in elo_table['round'].unique():
        elo_table = elo_table.drop(elo_table[elo_table['round']==r].index)
    for team in elo_table['team'].unique():
        if df[(df['Home'] == team)&(df['round']==r)]['Away'].shape[0]==0:
            continue
        
        if df[(df['Home'] == team)&(df['round']==r)]['Away'].values[0] == 'BYE':
            new_row = pd.DataFrame({'team':team,'round':r,'team_elo':elo_table[(elo_table['team']==team)&(elo_table['round']==r-1)]['team_elo'].values[0]}, index=[0])
            elo_table = pd.concat([elo_table, new_row])
            continue
        opp = df[(df['Home'] == team)&(df['round']==r)]['Away'].values[0]
        
        team_elo = elo_table[(elo_table['team'] == team)&(elo_table['round'] == r-1)]['team_elo'].values[0]
        opp_elo = elo_table[(elo_table['team'] == opp)&(elo_table['round'] == r-1)]['team_elo'].values[0]
        
        actual_outcome = df[(df['Home'] == team)&(df['round']==r)]['actual_outcome'].values[0]
        
        expected_outcome = get_expected_outcome(team_elo, opp_elo, 50, 10000)
        
        n_team_elo = team_elo + 100 * (actual_outcome - expected_outcome)
        n_opp_elo = opp_elo - 100 * (actual_outcome - expected_outcome)
        
        new_rows = pd.DataFrame({'team':[team, opp],'round':[r,r],'team_elo':[n_team_elo,n_opp_elo]}, index=[0,1])
        elo_table = pd.concat([elo_table, new_rows])
        
        df.loc[(df['Home']==team)&(df['round']==r),'expected_outcome'] = expected_outcome
        
elo_table = elo_table.sort_values(['round', 'team_elo'], ascending=[True,False]).reset_index(drop=True)
```
