```python
import os
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('/Users/caseyh/Desktop/footyfootyfooty/test.csv')
```

```python
data[data['year']==2023]
```

```python
data['venue_code'] = data['venue'].astype('category').cat.codes
data['round_code'] = data['round'].astype('category').cat.codes
```

```python
dic = {}
for i, team in enumerate(data['team'].unique()):
    dic[team] = i
for key in dic.keys():
    data.loc[data['team']==key,'team_code'] = dic[key]
for key in dic.keys():
    data.loc[data['opponent']==key,'opp_code'] = dic[key]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
data = data.dropna().reset_index(drop=True)
```

```python
train = data[data['year'] < 2023]
test = data[(data['year'] >= 2023)]
```

```python
predictors = ['year', 'at_home',
              'won_l5_form', 'loss_l5_form',
       'draw_l5_form', 'pf_l5_avg', 'pa_l5_avg', 'margin_l5_avg',
       'Disposals_l5_avg', 'Kicks_l5_avg', 'Handballs_l5_avg', 'Goals_l5_avg',
       'Behinds_l5_avg', 'Marks_l5_avg', 'Clearances_l5_avg', 'Tackles_l5_avg',
       'Contested Possessions_l5_avg', 'Uncontested Possessions_l5_avg',
       'Hit Outs_l5_avg', 'Contested Marks_l5_avg', 'Marks Inside 50_l5_avg',
       'Goal Assists_l5_avg', 'Frees_l5_avg', 'Frees Against_l5_avg',
       'Inside 50s_l5_avg', 'Rebounds_l5_avg', 'Clangers_l5_avg',
       'One Percenters_l5_avg', 'Bounces_l5_avg', 'position',
       'per', 'venue_code', 'team_code', 'opp_code']
```

```python
rf = RandomForestRegressor(n_estimators=100, min_samples_split=25, random_state=3)
```

```python
rf.fit(train[predictors], train['margin'])
```

```python
preds = rf.predict(test[predictors])
```

```python
test['prediction'] = preds
```

```python
a = test[['team','opponent','margin','prediction',
      'pf','pa','round']]
```
